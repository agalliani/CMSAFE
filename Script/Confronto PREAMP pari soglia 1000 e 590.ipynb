```python
import matplotlib.pyplot as plt
```

```python
corrente_preamp = [30,50,70] #microampere
```

```python
def plotConfronti(A_media_hit,B_media_hit, soglia):
    fig, ax = plt.subplots(figsize= (8,4))
    fig.suptitle("Confronto media hit soglia "+soglia+" elettroni (CD = 100fF)")
    
    ax.plot(corrente_preamp, A_media_hit, color = "blue", marker = "o")
    ax.plot(corrente_preamp, B_media_hit, color = "red",  marker = "o")
    ax.set_xlabel("Preamp Current (uA)")
    ax.set_ylabel("media hit")
    
    ax.legend([("RD53A"),("RD53B")])
    plt.savefig('Confronto pari soglia' + soglia, bbox_inches='tight')

    plt.show()
```

```python
A1000_media_hit = [15.09, 5.21, 0.71] #Capacit√† CD 100fF
B1000_media_hit = [130.41, 31.51, 42.61]
```

```python
plotConfronti(A1000_media_hit,B1000_media_hit, "1000")
```

```python
A590_media_hit = [622.46, 590.81, 714.97]
B590_media_hit = [928.12, 871.7, 817.62]
```

```python
plotConfronti(A590_media_hit,B590_media_hit, "590")
```
